```python
import pandas as pd
import os

columns = ['uf', 'cidade', 'zona', 'secao', 'cargo', 'candidato', 'votos']
df = pd.read_csv('./out.txt', sep=';', names=columns)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  
matplotlib.rcParams['figure.figsize'] = [12, 9]

votos_estado = df.groupby('uf').sum().sort_values('votos', ascending=False)
plt.bar(votos_estado.index, votos_estado.votos)
```
